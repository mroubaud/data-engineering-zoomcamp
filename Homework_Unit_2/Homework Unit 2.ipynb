```python
df_filter = pd.DataFrame()
for i in range(len(df)):
    row = date_series = df.loc[i,:]
    date_series = row["lpep_pickup_datetime"]
    date = pd.to_datetime(date_series)
    year = date.year
    month = date.month
    if ((year == 2020) & (10<=month<=12)):
        df_row = pd.DataFrame(row).T
        df_filter = pd.concat([df_filter, df_row])
```
